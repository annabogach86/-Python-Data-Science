```python
# С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pylab import rcParams
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(-3, 3, 51)
x
```

```python
# Создайте массивы y1, y2, y3, y4 по следующим формулам:
# y1 = x**2
# y2 = 2 * x + 0.5
# y3 = -3 * x - 1.5
# y4 = sin(x)
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
# Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.
# Figure с названием figи массив объектов Axes под названием ax,  
# причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. 
# В каждом графике массив x используется для координат по горизонтали.
# В левом верхнем графике для координат по вертикали используйте y1, в правом верхнем - y2, 
# в левом нижнем - y3, в правом нижнем - y4.
# Дайте название графикам: 'График y1', 'График y2' и т.д.
```

```python
figu, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_title('График y1')
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')
```

```python
# Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
```

```python
figu, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')
```

```python
# Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
```

```python
figu, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')
figu.set_size_inches(8, 6)
```

```python
# Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.
```

```python
figu, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')
figu.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3)
plt.subplots_adjust(hspace=0.3)
```
