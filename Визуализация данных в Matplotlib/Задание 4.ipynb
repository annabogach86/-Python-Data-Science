```python
#Данный датасет является примером несбалансированных данных, так как мошеннические операции с картами встречаются реже обычных.
#Импортируйте библиотеку Pandas, а также используйте для графиков стиль “fivethirtyeight”.
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pylab import rcParams
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.style.use('fivethirtyeight')
```

```python
#Посчитайте с помощью метода value_counts количество наблюдений для каждого значения целевой переменной Class 
#и примените к полученным данным метод plot, чтобы построить столбчатую диаграмму.
```

```python
DATASET_PATH = './creditcard.csv'
creditcard = pd.read_csv(DATASET_PATH)
creditcard.iloc[:,[0, 1, 2, 28, 29, 30]].head(10)
```

```python
class_list = creditcard['Class'].value_counts()
print(class_list)
```

```python
class_list.plot(kind='bar')
plt.show()
```

```python
# Затем постройте такую же диаграмму, используя логарифмический масштаб.
```

```python
class_list.plot(kind='bar', logy=True)
plt.show()
```

```python
#На следующем графике постройте две гистограммы по значениям признака V1 - одну для мошеннических транзакций (Class равен 1) 
# и другую - для обычных (Class равен 0).
```

```python
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
plt.hist(class0['V1'])
plt.hist(class1['V1'])
plt.show()
```

```python
#Подберите значение аргумента density так, чтобы по вертикали графика было расположено не число наблюдений,
# а плотность распределения. Число бинов должно равняться 20 для обеих гистограмм, 
# а коэффициент alpha сделайте равным 0.5, чтобы гистограммы были полупрозрачными и не загораживали друг друга. 
```

```python
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5)
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5)
plt.show()
```

```python
# Создайте легенду с двумя значениями: “Class 0” и “Class 1”.
```

```python
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1')
plt.legend()
plt.show()
```

```python
# Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного.
```

```python
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.legend()
plt.show()
```

```python
# Горизонтальной оси дайте название “V1”.
```

```python
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.xlabel('V1')
plt.legend()
plt.show()
```
